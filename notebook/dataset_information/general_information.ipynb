# General information regarding Cam-CAN data set

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style='ticks', palette='tab10')

from camcan.datasets import load_camcan_behavioural
```

```python
path_data = '/home/glemaitre/Documents/data/camcan/cc700-scored/total_score.csv'
path_participants_info = '/home/glemaitre/Documents/data/camcan/cc700-scored/participant_data.csv'

dataset = load_camcan_behavioural(path_data, path_participants_info)
```

### Information regarding the scores considered

```python
# Create a pandas DataFrame
X = pd.DataFrame(dataset.scores)
X['hand'][X['hand'].gt(0)] = 1
X['hand'][X['hand'].lt(0)] = 0
X['hand'][X['hand'] == 1] = 'Right-handed'
X['hand'][X['hand'] == 0] = 'Left-handed'
```

```python
sns.violinplot(x="hand", y="age", hue="gender_text", data=X, split=True, inner="quart")
```
